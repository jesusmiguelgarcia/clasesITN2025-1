```python
import gradio as gr
from transformers import pipeline

ner = pipeline("ner", grouped_entities=True, model="gunghio/distilbert-base-multilingual-cased-finetuned-conll2003-ner")

def extract_entities(text):
    entities = ner(text)
    result = [ [e['word'], e['entity_group'], f"{e['score']:.3f}"] for e in entities ]
    return f"Encontradas {len(entities)}",result

examples = [
    "Mi nombre es Dr. Garcia y trabajo en Universidad de la Sierra.",
    "La universidad de la Sierra se fundo en el 2002.",
    "My name is Dr. Garcia."
]
demo = gr.Interface(fn=extract_entities,
                    inputs=gr.Textbox(label="Texto en spanish or english"),
                    examples=examples,
                    outputs=[
                        gr.Textbox(label="Resultado"),
                        gr.Dataframe(headers=["Entidad", "Tipo", "Confianza"], label="Entidades")
                    ])
demo.launch()

```
